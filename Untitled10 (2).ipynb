```python
x=df['Salary'].mean()
df['Salary'].fillna(x,inplace=True)
df
```

```python
data=df[['Team','Position','Salary']]
data
```

```python
salary = data.groupby(['Team','Position'])['Salary'].sum().reset_index()
highest_spending = salary.loc[salary['Salary'].idxmax()]
highest_spending
```
